```python
import os
import csv
import numpy as np
import pandas as pd
import scipy.constants as sc
```

```python
data = pd.read_csv('data.csv')
data = data.drop('Unnamed: 0', axis=1)
data.head()
```

```python
temperature = 700
data['temp_check1'] = (temperature > data['Temperature range 1'].str.split().str[0].astype(float)) & (temperature < data['Temperature range 1'].str.split().str[-2].astype(float))
data['temp_check2'] = (temperature > data['Temperature range 2'].str.split().str[0].astype(float)) & (temperature < data['Temperature range 2'].str.split().str[-2].astype(float))
```

```python
data.head()
```

```python
data['a_2'] = data['a_2'].astype(float)
data['a_1'] = data['a_1'].astype(float)
data['a0'] = data['a0'].astype(float)
data['a1'] = data['a1'].astype(float)
data['a2'] = data['a2'].astype(float)
data['a3'] = data['a3'].astype(float)
data['a4'] = data['a4'].astype(float)
data['a5'] = data['a5'].astype(float)
data['a6'] = data['a6'].astype(float)

data['a_2.2'] = data['a_2.2'].astype(float)
data['a_1.2'] = data['a_1.2'].astype(float)
data['a0.2'] = data['a0.2'].astype(float)
data['a1.2'] = data['a1.2'].astype(float)
data['a2.2'] = data['a2.2'].astype(float)
data['a3.2'] = data['a3.2'].astype(float)
data['a4.2'] = data['a4.2'].astype(float)
data['a5.2'] = data['a5.2'].astype(float)
data['a6.2'] = data['a6.2'].astype(float)

T = temperature
```

```python
def calculate_H(T):
    val = sc.R*T*(-1*data['a_2']*T**-2 + data['a_1']*T**-1*np.log(T) + data['a0'] + 0.5*data['a1']*T + (data['a2']*T**2)/3 + 0.25*data['a3']*T**3 + 0.2*data['a4']*T**4 + data['a5']*T**-1)    

    return val

def calculate_H2(T):
    val = sc.R*T*(-1*data['a_2.2']*T**-2 + data['a_1.2']*T**-1*np.log(T) + data['a0.2'] + 0.5*data['a1.2']*T + (data['a2.2']*T**2)/3 + 0.25*data['a3.2']*T**3 + 0.2*data['a4.2']*T**4 + data['a5.2']*T**-1)    

    return val
```

```python
string_H = 'H_' + str(T)
string_S = 'S_' + str(T)
data[string_H] = data['temp_check1'].apply(lambda x: calculate_H(T) if x==True else calculate_H2(T))
```
